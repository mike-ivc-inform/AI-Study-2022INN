### Подготовка

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Подгружаем библиотеки и набор данных MINST dataset *(60 тыс изображений сканов черно-белых цифр от 0 до 0, написанных от руки. Наименование MNIST происходит от "modified NIST" - аббревиатура Национального института стандартов США, который собрал этот набор)*

*Графическй процессор не настроен, поэтому ниже появятся предупреждения - это не критично, работаем без настройки GPU*

```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.utils import np_utils
```

Загрузка данных с использованием **X** для выборок и **y** для маркировки *(Использование разного региста в обозначении выборки и маркировки - это такая конвенция)* 

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print("X_train shape", X_train.shape)
print("y_train shape", y_train.shape)
print("X_test shape", X_test.shape)
print("y_test shape", y_test.shape)
```

```python
%who
```

*это просто список переменных*

```python
plt.rcParams['figure.figsize'] = (9,9)

for i in range(9):
    plt.subplot(3,3,i+1)
    num = random.randint(0, len(X_train))
    
    plt.imshow(X_train[num], cmap='gray', interpolation='none')
    plt.title("Class {}, Image number: {}".format(y_train[num], num))
    
plt.tight_layout()
```

```python
def matprint(mat, fmt="g"):
    col_maxes = [max([len(("{:"+fmt+"}").format(x)) for x in col]) for col in mat.T]
    for x in mat:
        for i, y in enumerate(x):
            print(("{:"+str(col_maxes[i])+fmt+"}").format(y), end="  ")
        print("")

matprint(X_train[22660])
```

```python
plt.imshow(X_train[22660], cmap='gray', interpolation='none')
```

```python
X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train /= 255
X_test /= 255

print("Training matrix shape", X_train.shape)
print("Testing mayrix shape", X_test.shape)
```

```python
y_train[22660]
```

Конвертируем из числа в двоичный формат

```python
nb_classes = 10

Y_train = np_utils.to_categorical(y_train, nb_classes)
Y_test = np_utils.to_categorical(y_test, nb_classes)
```

```python
print(y_train[22660], Y_train[22660])
```

### Строим сеть:

```python
model = Sequential()
```

Добавляем первый слой:

```python
model.add(Dense(512, input_shape=(784,)))
```

Добавляем функцию активации:

```python
model.add(Activation('relu'))
```

Функция Dropout

```python
model.add(Dropout(0.2))
```

Второй блок кода:

```python
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.2))
```

Выходной слой:

```python
model.add(Dense(10))
```

```python
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, #60.000
          batch_size=128, epochs=10,
          verbose=1)
```

```python
score = model.evaluate(X_test, Y_test)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
predicted_classes = np.argmax(model.predict(X_test),axis=1)
correct_indices = np.nonzero(predicted_classes == y_test)[0]
incorrect_indices = np.nonzero(predicted_classes != y_test)[0]
```

```python
incorrect_indices[0]
```

```python
np.argmax(model.predict(X_test),axis=1).shape
```

```python
print('Real value is {}'.format(y_test[incorrect_indices[100]]))
print('Predicted value is {}'.format(predicted_classes[incorrect_indices[100]]))
plt.imshow(X_test[incorrect_indices[100]].reshape(28,28), cmap='gray', interpolation='none')
```

```python
plt.figure()
for i, correct in enumerate(correct_indices[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[correct].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[correct], y_test[correct]))
    
plt.tight_layout()
    
plt.figure()
for i, incorrect in enumerate(incorrect_indices[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[incorrect].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[incorrect], y_test[incorrect]))
    
plt.tight_layout()
```
